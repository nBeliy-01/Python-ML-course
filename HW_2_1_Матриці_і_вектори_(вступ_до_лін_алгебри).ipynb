### Завдання 1: Створення матриць та векторів
1. Створіть `numpy` вектор `vector` розмірністю 10 з випадковими дійсними значеннями в діапазоні від 0 до 1. Можна скористатись двома альтернативними варіантами:

  $$
  matrix = \begin{bmatrix}
  11 & 14 & 17 \\

  29 & 32 & 35

  $$

  будь-яким відомими Вам способом.

Виведіть на екран вектор і матрицю, а також їх розміри.


```python
import random
import numpy as np

vector = np.random.rand(10)
matrix = np.array([[11, 14, 17], [20, 23, 26], [29, 32, 35]])

print("Вектор: \n", vector, "\nРозмір вектора: ", vector.shape)
print("Матриця: \n", matrix, "\nРозмір матриці: ", matrix.shape)
```


### Завдання 2: Операції над векторами та матрицями
Виконайте кожну з зазначених операцій над зазначеними оʼєктами і виведіть результат на екран:
1. Додайте до кожного значення вектора `vector` число 10 і округліть кожне значення до 2 знаків після коми без використання циклів.


### Завдання 3: Множення матриць
1. Задані дві матриці `A` і `B`розмірностями 3x2 та 2х3. Виконайте їхнє множення, щоб отримати матрицю 2x2. Виведіть результуючу матрицю `result_matrix` на екран.

2. Заданий вектор `C` розмірністю 2х1. Подумайте спочатку, яку розмірність Ви отрмаєте, коли домножите `result_matrix` на вектор `C` і який результат буде. А потім виконайте множення і виведіть результат на екран.

3. Задана матриця `D` розмірністю 2х2. Подумайте спочатку, яку розмірність Ви отрмаєте, коли домножите `result_matrix` на матрицю `D` і який результат буде. А потім виконайте множення і виведіть результат на екран.

```python
A = np.array(
    [[0.18, 0.53],
     [0.75, 0.47],
     [0.35, 0.21]])

B = np.array(
    [[0.17, 0.58, 0.75],
     [0.38, 0.11, 0.15]])

C = np.array([[0], [1]])

D = np.array([[0,1], [1,0]])

display(A, B, C, D)
```

```python
# 1. Задані дві матриці A і Bрозмірностями 3x2 та 2х3. Виконайте їхнє множення, щоб отримати матрицю 2x2. Виведіть результуючу матрицю result_matrix на екран
result_matrix = np.dot(B, A)
print(result_matrix, '\n')

# 2. Заданий вектор C розмірністю 2х1. Подумайте спочатку, яку розмірність Ви отрмаєте, коли домножите result_matrix на вектор C і який результат буде. А потім виконайте множення і виведіть результат на екран.
# отримаю розмірність 2х1, оскільки result_matrix: 2x2 множиться на C: 2x1
print(np.dot(result_matrix, C), '\n')

# 3. Задана матриця D розмірністю 2х2. Подумайте спочатку, яку розмірність Ви отрмаєте, коли домножите result_matrix на матрицю D і який результат буде. А потім виконайте множення і виведіть результат на екран.
# отримаю розмірність 2х2, оскільки result_matrix: 2x2 множиться на D: 2x2

print(np.dot(result_matrix, D))
```


### Завдання 4: Транспонування матриць
1. Транспонуйте `result_matrix`.

3. Використайте операцію `numpy.expand_dims` над вектором `vector` з аргументом axis=1. Запишіть отриманий результат в змінну `column_vector` і виведіть на екран. Порівняйте з `vector`.
4. Зробіть транспонування `column_vector`, виведіть результат на екран разом з розмірностями отриманої структури.




```python
# 1. Транспонуйте result_matrix.
print(result_matrix.T, '\n')

# 2. Зробіть транспонування вектора vector і виведіть форму нової структури. Чи бачите ви зміну?
# не побачив зміни з не транспонованим вектором, такий спосіб транспонування можна використати у випадку якщо зробити vector.reshape(-1, 1), тоді зміна буде присутня
print(vector.T, '\n')

# 3. Використайте операцію numpy.expand_dims над вектором vector з аргументом axis=1. Запишіть отриманий результат в змінну column_vector і виведіть на екран. Порівняйте з vector.
# в даному випадку ми збільшили розмірність вектору, його розмір становить (10, 1), кожне число є масивом з єдиним елементом, тобто маємо масив з 10 масивів з 1 числом
column_vector = np.expand_dims(vector, axis=1)
print(column_vector)
print(column_vector.shape, '\n')

# 4. Зробіть транспонування column_vector, виведіть результат на екран разом з розмірностями отриманої структури.
# транспонування перетворило наші 10 масивів в 1 масив з масивом 10 чисел
print(column_vector.T)
print(column_vector.T.shape)
```

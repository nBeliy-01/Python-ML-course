1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
print("Функція: 10*x - 5")
print(f"Поіхдна функції 10*x - 5: {sp.diff('10*x - 5')}")
```

```python
def f(x):
    return 10*x - 5

def df(x):
    return 10
```

```python
_range = np.linspace(0, 10, 11)
plt.plot(_range, f(_range), marker='o', color='blue', label="Залежність оцінки від кількості годин")
plt.plot(_range, [df(_range)] * 11, color='red', label="Швидкість зміни оцінки (похідна)")
plt.legend()
plt.grid()
```

**Висновок:** в даному випадку оцінка лінійно залежна від кількості годин. Тобто що більша кількість годин витрачена на навчання, тим більшу оцінку можна отримати. Швидкість зміни оцінки в даному випадку є пряма яка показує, що кожна наступна оцінка збільшується з кількістю витрачених годин на 10.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
_range = np.linspace(0.1, 20, 100)

x = sp.symbols('x')

sin_x = sp.sin(x)
df_sin_x = sp.diff(sin_x, x)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(_range, [sin_x.subs(x, value) for value in _range], color='blue', label='sin(x)')
plt.plot(_range, [df_sin_x.subs(x, value) for value in _range], color='red', label='df sin(x)')
plt.grid()
plt.legend()

ln_x = sp.ln(x)
df_ln_x = sp.diff(ln_x, x)

plt.subplot(2, 2, 2)
plt.plot(_range, [ln_x.subs(x, value) for value in _range], color='blue', label='ln(x)')
plt.plot(_range, [df_ln_x.subs(x, value) for value in _range], color='red', label='df ln(x)')
plt.grid()
plt.legend()

plt.show()
```

На графіку зліва - функція `sin(x)` та її похідна `sin'(x) = cos(x)`<br>
На графіку справа - функція `ln(x)` та її похідна `ln'(x) = 1/x`<br>


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
_range = np.linspace(0, 100, 101)

def f(x):
    return x**2 - 5*x + 1

x = sp.symbols('x')
df_x = sp.diff('x^2 + 5*x + 1', x)

fig = plt.figure(figsize=(10, 6))

plt.plot(_range, f(_range), label='Функція загальних витрат компанії', color='blue')
plt.plot(_range, [df_x.subs(x, value) for value in _range], label='Похідна функції, швидкість зміни витрат', color='red')
plt.legend()
plt.grid()

print(f'Похідна функції f(x) = x^2 - 5x + 1: {df_x}')
```
